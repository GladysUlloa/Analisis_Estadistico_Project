### **1️⃣ Contexto del Estudio**

El dataset Iris, introducido por el estadístico y genetista Ronald Fisher en 1936, es uno de los conjuntos de datos más utilizados en estadística y machine learning.

Contiene 150 observaciones de flores clasificadas en tres especies:

* Setosa
* Versicolor
* Virginica

Cada observación incluye cuatro variables cuantitativas:

* Sepal length (cm)
* Sepal width (cm)
* Petal length (cm)
* Petal width (cm)

## **🎯 Objetivo del Análisis**

Analizar la estructura de dependencia lineal entre variables mediante la matriz de correlación y evaluar implicaciones para modelado estadístico y aprendizaje automático.

### **2️⃣ Metodología**

El análisis se desarrollará en cinco etapas:

1. Carga y exploración inicial
2. Cálculo de la matriz de correlación
3. Visualización

5. Implicaciones para modelado

### **3️⃣ Código Completo en Python**

### **🔹 3.1 Carga de datos**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar dataset
iris = load_iris()

# Crear DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Añadir variable objetivo
df["species"] = iris.target

df.head()
```

### **🔹 3.2 Estadísticos descriptivos**

```python
df.describe()
```

### **🔹 3.3 Matriz de correlación**

```python
corr_matrix = df.drop(columns=["species"]).corr()
corr_matrix
```

### **🔹 3.4 Visualización**

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Matriz de Correlación - Dataset Iris")
plt.show()
```

### **5️⃣ Evaluación de Multicolinealidad (VIF)**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df.drop(columns=["species"]).values

vif_data = pd.DataFrame()
vif_data["Variable"] = df.drop(columns=["species"]).columns
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

vif_data
```

### **6️⃣ Interpretación Estadística**

La matriz de correlación estimada es una aproximación empírica de la matriz:

R = Corr(X)

Donde X es el vector aleatorio multivariado:

X = (X₁, X₂, X₃, X₄)

Esta matriz:
* Es simétrica.
* Es positiva semidefinida.
* Describe la estructura de segundo orden.

En términos prácticos, estamos evaluando la dependencia lineal entre componentes del sistema morfológico floral.

### **7️⃣ Implicaciones para Machine Learning**
🔹 En regresión lineal

Alta correlación → coeficientes inestables.

🔹 En PCA

Las variables altamente correlacionadas dominarán el primer componente principal.

🔹 En clustering

Podrían sesgar distancias euclidianas.

🔹 En selección de variables

Podría justificarse eliminar una variable redundante.
